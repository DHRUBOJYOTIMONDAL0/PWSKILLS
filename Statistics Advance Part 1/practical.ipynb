```python
#3.  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pdf(x, p):
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        return 0
p = 0.7

x1 = 1
x0 = 0

print("P(X = 1) =", bernoulli_pdf(x1, p))
print("P(X = 0) =", bernoulli_pdf(x0, p))

```

```python
#6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
#uniform distribution.
import numpy as np
import matplotlib.pyplot as plt

values = np.array([1, 2, 3, 4, 5])
n = len(values)
pmf = np.ones(n) / n
cdf = np.cumsum(pmf)

plt.step(values, cdf, where='post')
plt.xlabel("Discrete Random Variable (X)")
plt.ylabel("CDF  P(X  x)")
plt.title("CDF of Discrete Uniform Distribution")
plt.ylim(0, 1.1)
plt.grid(True)
plt.show()
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a given dataset.

    Parameters:
    data : list or numpy array

    Returns:
    numpy array of Z-scores
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores


data = np.array([10, 12, 15, 18, 20, 22, 25])

z_scores = calculate_z_scores(data)

print("Data:", data)
print("Z-scores:", z_scores)
plt.plot(data, z_scores, marker='o')
plt.axhline(0)
plt.xlabel("Data Values")
plt.ylabel("Z-score")
plt.title("Z-scores of Dataset")
plt.grid(True)
plt.show()
```

```python
#10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt

population_samples = 10000
sample_size = 30
num_samples = 1000

population = np.random.uniform(0, 10, population_samples)

sample_means = []
for _ in range(num_samples):
    sample = np.random.choice(population, sample_size)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30, density=True, edgecolor='black')
plt.xlabel("Sample Mean")
plt.ylabel("Probability Density")
plt.title("Central Limit Theorem Demonstration (Uniform Distribution)")
plt.show()
```

```python
#12. Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1).
import numpy as np
import matplotlib.pyplot as plt

def standard_normal_distribution():
    x = np.linspace(-4, 4, 1000)

    pdf = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

    plt.plot(x, pdf)
    plt.xlabel("x")
    plt.ylabel("Probability Density")
    plt.title("Standard Normal Distribution ( = 0,  = 1)")
    plt.grid(True)
    plt.show()

standard_normal_distribution()
```

```python
#13.  Generate random variables and calculate their corresponding probabilities using the binomial distribution.
import numpy as np
from math import comb

n = 10
p = 0.5
samples = 5
random_values = np.random.binomial(n, p, samples)

print("Random Variables:", random_values)

print("\nCorresponding Probabilities:")
for k in random_values:
    probability = comb(n, k) * (p ** k) * ((1 - p) ** (n - k))
    print(f"P(X = {k}) = {probability}")
```

```python
#14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
# distribution.
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-score
def calculate_z_score(x, mean, std):
    return (x - mean) / std

data = np.array([10, 12, 15, 18, 20, 22, 25])

x = 18
mean = np.mean(data)
std = np.std(data)

z = calculate_z_score(x, mean, std)

print("Mean =", mean)
print("Standard Deviation =", std)
print("Z-score for x =", x, "is", z)

z_values = np.linspace(-4, 4, 1000)
pdf = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * z_values**2)

plt.plot(z_values, pdf, label="Standard Normal Distribution")
plt.axvline(z, linestyle='--', label=f"Z = {z:.2f}")
plt.xlabel("Z")
plt.ylabel("Probability Density")
plt.title("Z-score Comparison with Standard Normal Distribution")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm
data = np.array([52, 55, 50, 54, 53, 56, 51, 52, 54, 55])
mu_0 = 50
sigma = 4
alpha = 0.05

# Sample statistics
sample_mean = np.mean(data)
n = len(data)
z = (sample_mean - mu_0) / (sigma / np.sqrt(n))
z_critical = norm.ppf(1 - alpha / 2)

print("Sample Mean =", sample_mean)
print("Z-statistic =", z)
print("Critical Z-value =", z_critical)

if abs(z) > z_critical:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```

```python
#16. Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
from scipy.stats import norm
data = np.array([52, 55, 50, 54, 53, 56, 51, 52, 54, 55])

confidence_level = 0.95
sigma = 4

mean = np.mean(data)
n = len(data)
z = norm.ppf((1 + confidence_level) / 2)
margin_error = z * (sigma / np.sqrt(n))
lower = mean - margin_error
upper = mean + margin_error

print("Sample Mean =", mean)
print("95% Confidence Interval = (", lower, ",", upper, ")")
```

```python
#17.  Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
import numpy as np
from scipy.stats import t

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=30)

mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

confidence_level = 0.95
alpha = 1 - confidence_level

t_critical = t.ppf(1 - alpha/2, df=n-1)

margin_error = t_critical * (std / np.sqrt(n))


lower = mean - margin_error
upper = mean + margin_error
print("Sample Mean =", mean)
print("95% Confidence Interval = (", lower, ",", upper, ")")
```

```python
#18. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.
import numpy as np
import matplotlib.pyplot as plt

mean = 0
std_dev = 1

x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

plt.plot(x, pdf)
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.title("Probability Density Function of Normal Distribution")
plt.grid(True)
plt.show()
```

```python
#21.  Write a Python program to compare the standard deviations of two datasets and visualize the difference.
import numpy as np
import matplotlib.pyplot as plt

data1 = np.array([10, 12, 14, 16, 18, 20])
data2 = np.array([10, 11, 12, 13, 14, 15])

std1 = np.std(data1, ddof=1)
std2 = np.std(data2, ddof=1)

print("Standard Deviation of Dataset 1:", std1)
print("Standard Deviation of Dataset 2:", std2)

datasets = ['Dataset 1', 'Dataset 2']
std_values = [std1, std2]

plt.bar(datasets, std_values)
plt.ylabel("Standard Deviation")
plt.title("Comparison of Standard Deviations")
plt.show()
```

```python
#22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.
import numpy as np


np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)

data_range = np.max(data) - np.min(data)

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)

IQR = Q3 - Q1
print("Range =", data_range)
print("Q1 (25th percentile) =", Q1)
print("Q3 (75th percentile) =", Q3)
print("Interquartile Range (IQR) =", IQR)
```

```python
#23.  Implement Z-score normalization on a dataset and visualize its transformation.
import numpy as np
import matplotlib.pyplot as plt

data = np.array([10, 12, 15, 18, 20, 22, 25])

mean = np.mean(data)
std = np.std(data)

z_scores = (data - mean) / std

print("Original Data:", data)
print("Z-score Normalized Data:", z_scores)

plt.figure(figsize=(8, 4))

plt.plot(data, marker='o', label='Original Data')

plt.plot(z_scores, marker='o', label='Z-score Normalized Data')

plt.axhline(0, linestyle='--', linewidth=1)
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Z-score Normalization Transformation")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
#distribution
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis():
    np.random.seed(42)
    data = np.random.normal(loc=50, scale=10, size=1000)

    skewness = skew(data)
    kurt = kurtosis(data)

    return skewness, kurt


skewness, kurt = calculate_skewness_kurtosis()

print("Skewness:", skewness)
print("Kurtosis:", kurt)
```
