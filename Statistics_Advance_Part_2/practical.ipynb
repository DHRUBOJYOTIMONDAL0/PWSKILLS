```python
#6. Question 6: Write a Python program to generate a binomial distribution with n=10 and
#p=0.5, then plot its histogram.
#(Include your Python code and output in the code box below.)
#Hint: Generate random number using random function
# Ans- A binomial distribution models the number of successes in a fixed number of independent trials, where each trial has the same probability of success.
# Number of trials, n = 10
#Probability of success, p = 0.5
#We generate random values using Pythonâ€™s random function and then plot a histogram to visualize the distribution.

# Code-
import random
import matplotlib.pyplot as plt

n = 10
p = 0.5
samples = 1000

binomial_data = []

for _ in range(samples):
    successes = 0
    for _ in range(n):
        if random.random() < p:
            successes += 1
    binomial_data.append(successes)

plt.hist(binomial_data, bins=range(0, n+2), edgecolor='black')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.show()
```

```python
#7. Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
#Python. Show the Python code and interpret the results.
#sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
#50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
#50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
#(Include your Python code and output in the code box below.

#Ans-
import math

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
mu = 50

# Calculations
n = len(sample_data)
sample_mean = sum(sample_data) / n
variance = sum((x - sample_mean) ** 2 for x in sample_data) / n
std_dev = math.sqrt(variance)

# Z-statistic
z = (sample_mean - mu) / (std_dev / math.sqrt(n))

# Output
print("Sample Size (n):", n)
print("Sample Mean:", round(sample_mean, 4))
print("Standard Deviation:", round(std_dev, 4))
print("Z-statistic:", round(z, 4))

```

```python
#8. Question 8: Write a Python script to simulate data from a normal distribution and
#calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
#(Include your Python code and output in the code box below.)

#Ans-
import random
import math
import matplotlib.pyplot as plt
random.seed(1)
n = 100
mu = 50
sigma = 5

data = [random.gauss(mu, sigma) for _ in range(n)]

#Calculate sample mean and standard deviation
sample_mean = sum(data) / n
variance = sum((x - sample_mean) ** 2 for x in data) / n
std_dev = math.sqrt(variance)

#Calculate 95% confidence interval
z_value = 1.96
margin_error = z_value * (std_dev / math.sqrt(n))

lower_bound = sample_mean - margin_error
upper_bound = sample_mean + margin_error

print("Sample Mean:", round(sample_mean, 4))
print("Standard Deviation:", round(std_dev, 4))
print("95% Confidence Interval: (",
      round(lower_bound, 4), ",", round(upper_bound, 4), ")")

#histogram
plt.hist(data, bins=10, edgecolor='black')
plt.axvline(sample_mean, linestyle='dashed', linewidth=2)
plt.title("Normal Distribution (Simulated Data)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#9. Question 9: Write a Python function to calculate the Z-scores from a dataset and
#visualize the standardized data using a histogram. Explain what the Z-scores represent
#in terms of standard deviations from the mean.
#(Include your Python code and output in the code box below.)

#Ans-
import math
import matplotlib.pyplot as plt
def calculate_z_scores(data):
    n = len(data)
    mean = sum(data) / n
    variance = sum((x - mean) ** 2 for x in data) / n
    std_dev = math.sqrt(variance)

    z_scores = [(x - mean) / std_dev for x in data]
    return z_scores, mean, std_dev

# Sample dataset
data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5]

#Z-scores
z_scores, mean, std_dev = calculate_z_scores(data)

print("Mean:", round(mean, 4))
print("Standard Deviation:", round(std_dev, 4))
print("Z-scores:")
for z in z_scores:
    print(round(z, 3))

#histogram of Z-scores
plt.hist(z_scores, bins=10)
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Histogram of Z-scores (Standardized Data)")
plt.show()
```
