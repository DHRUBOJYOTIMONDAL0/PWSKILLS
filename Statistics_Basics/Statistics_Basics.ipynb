Height (0 cm, 50 cm, 100 cm)

4. What is a box plot, and what can it tell you about the distribution of data?
Ans- Box Plot Overview
A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a data set. It visually displays the central tendency, variability, and skewness of the data.

Key Components of a Box Plot
Box: The main part of the plot, which represents the interquartile range (IQR). The box extends from the first quartile (Q1) to the third quartile (Q3).

Median Line: A line inside the box that shows the median (Q2) of the data set.

Whiskers: Lines extending from the box to the minimum and maximum values within a specified range (usually 1.5 times the IQR).

Outliers: Individual points outside the whiskers, representing data points that are unusually high or low.

What a Box Plot Tells You
Central Tendency: The median line inside the box shows the middle value of the data set.

Spread and Variability: The length of the box (IQR) indicates the range within which the middle 50% of the data lies. The whiskers show the overall spread of the data.

Skewness: The position of the median line within the box indicates skewness. If the median is closer to the lower or upper quartile, the data is skewed.

Outliers: Points outside the whiskers indicate outliers, which can provide insights into unusual observations or potential errors in the data.

Example
Consider a data set of student test scores: [55, 60, 65, 70, 75, 80, 85, 90, 95, 100].

Q1 (first quartile): 65

Median (Q2): 75

Q3 (third quartile): 85

Minimum (excluding outliers): 55

Maximum (excluding outliers): 100

In this box plot:

The box spans from 65 to 85.

The median line is at 75.

Whiskers extend to 55 and 100.

There are no outliers.

5.  Discuss the role of random sampling in making inferences about populations.
Ans- Role of Random Sampling in Making Inferences About Populations
Random sampling is a fundamental technique used in statistics to make inferences about a population based on a subset of that population. It involves selecting a sample in such a way that every individual in the population has an equal chance of being chosen. This method ensures that the sample is representative of the entire population, which is crucial for accurate and reliable statistical inferences.
Representativeness:-
Definition: A representative sample accurately reflects the characteristics of the population from which it is drawn.
Importance: By ensuring that the sample is representative, random sampling helps in generalizing the findings from the sample to the entire population.
Minimizing Bias:-
Definition: Bias occurs when the sample does not accurately reflect the population, leading to skewed results.
Importance: Random sampling reduces selection bias by giving each member of the population an equal opportunity to be included in the sample. This enhances the validity of the results.
Statistical Validity:-
Definition: Statistical validity refers to the extent to which the conclusions drawn from the data are accurate and reliable.
Importance: Random sampling allows for the use of statistical methods to make valid inferences about the population. It provides the basis for calculating confidence intervals and performing hypothesis tests.
Precision of Estimates:
Definition: Precision refers to the closeness of the estimated values to the true population values.
Importance: Random sampling increases the precision of population estimates by reducing sampling error. Larger random samples tend to provide more precise estimates.
Application in Various Fields:-
Definition: Random sampling is used in numerous fields such as healthcare, market research, social sciences, and politics.
Importance: In healthcare, random sampling is used in clinical trials to test the efficacy of new treatments. In market research, it helps businesses understand consumer preferences. In social sciences, it aids in studying human behavior and social trends. In politics, it is used in opinion polls to gauge public sentiment.

6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?
 Ans- Concept of Skewness
Definition: Skewness refers to the asymmetry in the distribution of data, indicating whether data points are more concentrated on one side of the mean than the other.
Types of Skewness
Positive Skewness (Right-Skewed)
Definition: The tail on the right side is longer or fatter. Most data points are on the left.
Example: Income distribution where a majority earn less and a few earn significantly more.
Negative Skewness (Left-Skewed)
Definition: The tail on the left side is longer or fatter. Most data points are on the right.
Example: Age at retirement where most retire around a certain age, with a few retiring much earlier.
No Skewness (Symmetrical)
Definition: Data is evenly distributed around the mean, with no skewness. The distribution is symmetrical.
Example: Heights of adult men in a given population.

7. What is the interquartile range (IQR), and how is it used to detect outliers?
Interquartile Range (IQR)
The Interquartile Range (IQR) is a measure of statistical dispersion, which is the spread of the data points in a data set. It is the range between the first quartile (Q1) and the third quartile (Q3).
Calculation:-
First Quartile (Q1): The median of the lower half of the data set (25th percentile).
Third Quartile (Q3): The median of the upper half of the data set (75th percentile).
IQR: The difference between Q3 and Q1.
Formula: IQR = Q3 - Q1
Example
Consider the data set: [3, 7, 8, 12, 14, 18, 21, 24, 30]
Q1: The median of the first half (excluding the overall median): [3, 7, 8, 12] ‚Üí Q1 = 7.5
Q3: The median of the second half (excluding the overall median): [18, 21, 24, 30] ‚Üí Q3 = 22.5
IQR: Q3 - Q1 = 22.5 - 7.5 = 15
Detecting Outliers Using IQR
Outliers are data points that fall significantly outside the range of the majority of the data. They can be detected using the IQR method.
Calculation of Outliers:-
Lower Bound: Q1 - 1.5 * IQR
Upper Bound: Q3 + 1.5 * IQR
Example:-
Lower Bound: 7.5 - (1.5 * 15) = 7.5 - 22.5 = -15
Upper Bound: 22.5 + (1.5 * 15) = 22.5 + 22.5 = 45
In this example, any data point below -15 or above 45 would be considered an outlier.

9.Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule)
Ans- Properties of the Normal Distribution
Symmetry: The normal distribution is symmetric around its mean. This means that the left side of the distribution is a mirror image of the right side.
Bell-Shaped Curve: The graph of the normal distribution forms a bell-shaped curve, with the highest point at the mean and tails that extend infinitely in both directions.
Mean, Median, and Mode: In a normal distribution, the mean, median, and mode are all equal and located at the center of the distribution.
Asymptotic Nature: The tails of the normal distribution curve approach the horizontal axis but never touch it. This means that there are always some values, however extreme, within the distribution.
Empirical Rule (68-95-99.7 Rule)
The empirical rule provides a way to understand the spread of data in a normal distribution. According to this rule:
68% of Data: Approximately 68% of the data falls within one standard deviation (œÉ) of the mean (Œº). This means that most data points are relatively close to the mean.
